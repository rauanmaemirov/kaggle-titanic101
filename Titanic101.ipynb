### Cleaning data

Basically, we are going to fill out all our missing values with median of corresponding column.  
You could try to meticulously fill them out by group medians (or means).

Here's the list of ideas for cleaning data and *feature generation*:
- Fill missing age values by median age within same gender (male/female) or title (mr, mrs, miss, master)
- Fill missing fare values with median fare value for appropriate passenger class
- Calculate true values for fare. Some tickets indicate money payed for all family/group members, `Cabin` values for those tickets might look like 'C24 C26 C28', and all group members will have the same `Fare` and `Cabin` values.
- Separate `SibSp` column. Underages cannot have spouse.
- Separate `Parch` column.
- Guess missing age by `SibSp` and `Parch`. Infants cannot have `Parch` values higher than 2 or, probably, equal to 0.
- Try to add `Rank` variable for military and aristocratic ranks. (Countess, Sir, Col, Capt, etc in `Name`)


````python
df.Age.fillna(df.Age.median(), inplace=True)
df.Fare.fillna(df.Fare.median(), inplace=True)

# Set to port, where the most of passengers were embarked
max_embarked = df.groupby('Embarked').count()['PassengerId']
df.Embarked.fillna(max_embarked[max_embarked == max_embarked.max()].index[0], inplace=True)

df.dropna(inplace=True)
````

### Extracting features

Before running our classifier, we have last preparations to do. We need to choose which variables to use as features. I experimented with all meaningful variables and even tried to use generated by me, such as age categories or cabin sections, but with no luck.

Some algorithm implementations don't understand categorical variables, we need to encode them somehow. We are going to use nice `factorize` method built-in to Pandas. It simply puts column's unique values into a list and returns corresponding index.

````python
df['Sex_vect'], _ = pd.factorize(df['Sex'])

features = np.array(['Age', 'Fare', 'Sex_vect', 'Pclass', 'SibSp', 'Parch'])

x_train = df[features]
y_train = df['Survived']
````

If your algorithm of choice works better with sparse data, there's also [get_dummies](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.core.reshape.get_dummies.html) method. Scikit-learn has nice features like [OneHotEncoder](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html), [LabelEncoder](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html).

Using [Patsy](http://patsy.readthedocs.org/en/latest/categorical-coding.html) is another way to extract features, which appears to be the easiest one.

```
formula = 'Age + Fare + C(Sex) + Pclass + SibSp + Parch - 1'

```

It encodes our categorical variables in one-of-K scheme.

### Firing up our algorithm

I chose *Random Forrest*, but you might want to use Support Vector Machines or K-Nearest Neighboors. Logistic regression is also fine.

(Try to tune classifier [parameters](http://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html))

````python
clf = RandomForestClassifier(n_estimators=80, criterion='entropy', max_depth=5)

cv = cross_validation.KFold(len(df), n_folds=5, indices=False)

results = []

for traincv, testcv in cv:
    probas = clf.fit(x_train[traincv], y_train[traincv]).predict_proba(x_train[testcv])
    results.append(clf.score(x_train, y_train))

print('Accuracy: ' + str(np.array(results).mean()))

# Retrain classifier on entire set
clf.fit(x_train, y_train);
````

### Feature importance

For this particular algorithm, scikit-learn can show which features of the classifier are the most important ones.

Sometimes it comes in really handy. For instance, during first evaluation of my model `Fare` column ended up to be the most important, and `feature importance` tool helped me to realise early on that the model was highly overfitted.

````python
feature_importance = clf.feature_importances_
feature_importance = 100.0 * (feature_importance / feature_importance.max())

sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + .5

plt.barh(pos, feature_importance[sorted_idx], align='center')
plt.yticks(pos, df[features].columns[sorted_idx])
plt.title('Variable Importance');
````

### Predicting our test set and making a submission

````python
# Prepare test set
test_df = pd.read_csv('data/test.csv')
test_df.Age.fillna(test_df.Age.median(), inplace=True)
test_df.Fare.fillna(test_df.Fare.median(), inplace=True)

max_embarked = test_df.groupby('Embarked').count()['PassengerId']
test_df.Embarked.fillna(max_embarked[max_embarked == max_embarked.max()].index[0], inplace=True)

test_df['Sex_vect'], _ = pd.factorize(test_df['Sex'])
test_df.fillna(0, inplace=True)

test_df['Survived'] = clf.predict(test_df[features])

test_df.to_csv('data/submission.csv', cols=['PassengerId', 'Survived'], index=False, float_format="%d")
````
